### Features with high null ratio

```python
def plot_missing_values(X):
    """ For each column with missing values plot ratio that is missing."""
    data = [(col, df[col].isnull().sum() / len(df)) 
            for col in df.columns if df[col].isnull().sum() > 0]
    col_names = ['column', 'percent_missing']
    missing_df = pd.DataFrame(data, columns=col_names).sort_values('percent_missing')
    pylab.rcParams['figure.figsize'] = (15, 8)
    missing_df.plot(kind='barh', x='column', y='percent_missing'); 
    plt.title('Ratio of missing values in columns');
```

### Imputing

```python
# Numerical features selection. There is no ordinal variable in the dataset.
# Integers determined by chemical analytical resolution (i.e. Cr)

num_feat = X.select_dtypes(exclude=['object'])
num_feat.columns
```

```python
# Categorical features selection

cat_feat = X.select_dtypes(include=['object'])
cat_feat = pd.DataFrame(cat_feat)
cat_feat
```

```python
is_null_sum = X.isnull().sum()
is_null_sum
```

```python
# List of categorical features with missing values

cat_feat_miss = [var for var in cat_feat if is_null_sum[var]>0]
cat_feat_miss
```

```python
# List of numerical features with missing values

num_feat_miss = [var for var in num_feat if is_null_sum[var]>0]
num_feat_miss
```

```python
# Median as imputer was tested. Mean is the best strategy

model_imputer = SimpleImputer()

model_imputer.fit(X[num_feat_miss])

X[num_feat_miss] = model_imputer.transform(X[num_feat_miss])
```

```python
# KNN as imputer. Mean is the best strategy

# model_imputer = KNNImputer(n_neighbors=5)
# X[num_feat_miss] = model_imputer.fit_transform(X[num_feat_miss])
```

```python
model_imputer = SimpleImputer(strategy="most_frequent")

model_imputer.fit(X[cat_feat_miss])

X[cat_feat_miss] = model_imputer.transform(X[cat_feat_miss])
```

```python
# Negative root-mean-square error (NRMSE) by algorithm

count = 0
for model in models:
    scores = cross_val_score(model, final_dataset, y, cv=10, scoring='neg_root_mean_squared_error')
    score = scores.mean()
    print(f"NRMSE for {algorithm[count]} is {score:.3f}")
    count += 1
```
